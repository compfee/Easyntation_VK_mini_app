```python
import re
def remove_punct(text):
    """
        Remove the punctuation
    """
#     return re.sub(r'[]!"$%&\'()*+,./:;=#@?[\\^_`{|}~-]+', "", text)
    s = string.punctuation + '•»«–�−' + string.digits
    clean_text = re.sub(r"[^А-Яа-я\s]+", "", text)
    return clean_text.translate(str.maketrans('', '', s))
```

```python
from nltk.stem.snowball import SnowballStemmer 

def snowball_stemmer(lst):
    """
        Stem words in list of tokenized words with SnowballStemmer
    """
    stemmer = SnowballStemmer("russian")
    s = []
    ll=0
    for l in lst:
        
        stems = [stemmer.stem(i) for i in l]
        s.append(stems)
        ll+=len(stems)
    print(ll)
    
    return s, ll
```

```python
%time 

df['snowball_stemmer'], df['snowball_stemmer_len'] = df['stopwords_removed'].apply(lambda x: snowball_stemmer(x))
df.head()
```
